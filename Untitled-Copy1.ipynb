```python
## filling missing values with mean of their columns
data.fillna(data.mean(),inplace=True)
```

```python
## in data date is object lets convert it into pd datetime 
data["Date"]=pd.to_datetime(data["Date"])
```

```python
## importing news hedlines dataset
news_headlines=pd.read_csv(r"C:\Users\Gaonkar\Desktop\Ml+Dl+Rl_Projects\Ml\Sparks_Foundation\Task7 Stock and Sentiment\india-news-headlines.csv")
```

```python
news_headlines.shape
```

```python
news_headlines.head()
```

```python
# coverting publish_date to pd datetime format
year=[news_headlines["publish_date"].apply(lambda x : str(x)[:4])]
month=[news_headlines["publish_date"].apply(lambda x : str(x)[4:6])]
day=[news_headlines["publish_date"].apply(lambda x : str(x)[6:])]
year=np.array(year)
month=np.array(month)
day=np.array(day)
date=pd.DataFrame({"year":year[0],"month":month[0],"day":day[0]})
date=pd.to_datetime(date)
```

```python
date.head()
```

```python
news_headlines["date"]=date
news_headlines.drop(["publish_date","headline_category"],axis=1,inplace=True)
news_headlines.head()
```

```python
news_headlines.nunique()
```

```python
## seems like unique headlines are less than actual shape of dataset
news_headlines.drop_duplicates(inplace=True)
```

```python
news_headlines = news_headlines.groupby(['date'])['headline_text'].apply(lambda x: ','.join(x)).reset_index()
news_headlines.set_index('date', inplace= True)
news_headlines.head()
```

```python
## function to find sentiment polarity using textblob
def sentiment(headline):
    result=TextBlob(headline)
    return result.sentiment.polarity
```

```python
news_headlines["Sentiment"]=[sentiment(headline) for headline in news_headlines["headline_text"]]
```

```python
news_headlines.sort_index(ascending=True,axis=0,inplace=True)
news_headlines.head()
```

```python
positive_headline=news_headlines[news_headlines["Sentiment"]>0]
negative_headline=news_headlines[news_headlines["Sentiment"]<0]
neutral_headline=news_headlines[news_headlines["Sentiment"]==0]
per_pos=len(positive_headline)*100/len(news_headlines)
per_neg=len(negative_headline)*100/len(news_headlines)
per_neu=len(neutral_headline)*100/len(news_headlines)
print(f'Positive_headline = {per_pos}%')
print(f'negative_headline = {per_neg}%')
print(f'neutral_headline = {per_neu}%')
```

```python
data=data.set_index("Date")
data.sort_index(ascending=True,axis=0,inplace=True)
```

```python
data.head()
```

```python
## concatenating the dfs
data_comb=pd.concat([data,news_headlines],axis=1)
```

```python
data_comb.head()
```

```python
data.describe()
```

```python
data_comb.drop(["headline_text"],axis=1,inplace=True)
data_comb.fillna(data_comb.mean(),inplace=True)
pd.isnull(data_comb).sum()
```

```python
data_comb.shape
```

```python
## splitting data into train and test split 98% and 2% split
X=data_comb.drop(["Close","Volume"],axis=1)
y=data_comb["Close"]
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.02,random_state=43)
```

```python
# scaling the data
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
y_train=np.array(y_train).reshape(len(y_train),1)
y_test=np.array(y_test).reshape(len(y_test),1)
sc1=StandardScaler()
sc1=StandardScaler()
y_train=sc1.fit_transform(y_train)
y_test=sc1.transform(y_test)
```

```python
error=[]
for i in range(1,5):
    knn=KNeighborsRegressor(i)
    knn.fit(x_train,y_train)
    score=cross_val_score(knn, x_train, y_train, cv = 5, scoring = 'neg_mean_squared_error')
    print("cross_val_score of {} neigbors is {}".format(i, -1*score.mean()))
    y_pred=knn.predict(x_test)
    mse=mean_squared_error(y_test,y_pred)
    error.append(np.sqrt(mse))  
```

```python
plt.plot(range(1,5),error)
plt.xlabel("K values")
plt.ylabel("MSE")
```

```python
## training classfier with three neighbors seems error is very high 
## dont compare the above mse there data was scale ,this error is of unscale y
knn=KNeighborsRegressor(3)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
#============ unscaling ===============
y_te=sc1.inverse_transform(y_test)
y_pr=sc1.inverse_transform(y_pred)
#=====================================
mse=mean_squared_error(y_te,y_pr)
mse
```

```python
# though the graph look overlapped completely it is not overlapping when we zoom the graph we will see the error
## we can tell this because mse is high .
plt.figure(figsize=(16,10))
plt.plot(y_pr, label="Predicted Close Price")
plt.plot([row[0] for row in y_te], label="Testing Close Price")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, shadow=True, ncol=2)
plt.show()
```

```python
## randomforest peforms exceptionally well than knn 
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
y_te=sc1.inverse_transform(y_test)
y_pr=sc1.inverse_transform(y_pred)
mse=mean_squared_error(y_te,y_pr)
mse
```

```python
## mse proves that rf is performing well than knn
plt.figure(figsize=(16,10))
plt.plot(y_pr, label="Predicted Close Price")
plt.plot(y_te, label="Testing Close Price")
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, shadow=True, ncol=2)
plt.show()
```
